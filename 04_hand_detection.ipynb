> verify that hands are detected by the hand detection model

```python
import os
from asl_detection.detection.utils import draw_boxes, load_model
```

```python
hand_detector_path = '/home/nic/repos/darknet/backup/asl/asl-yolov3_best.weights'
hand_detector = load_model(hand_detector_path, weights=True)

files_path = '/home/nic/data/asl/imgs'
files = os.listdir(files_path)
files = [os.path.join(files_path, file) for file in files][:5]
save_dir = './results/detection'
os.makedirs(save_dir, exist_ok=True)
```

```python
draw_boxes(hand_detector,
                     files,
                     figsize=(10, 10),
                     save_dir=save_dir)
```
